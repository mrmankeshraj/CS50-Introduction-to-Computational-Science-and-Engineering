```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data For plotting
x = np.linspace(0.0, 1.0, 31)
f1 = x
f2 = x**2

# Plotting
fig = plt.figure()

plt.plot(x, f1, color="r", marker="o", linestyle="-.", label="f(x)=x")
plt.plot(x, f2, color="b", marker="x", linestyle="-.", label="f(x)=x^2")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Our first plot!")

# Draw the grid on the figure
plt.grid()

# Create the legend
plt.legend()

# Save the file to a file you can send to your friends
fig.savefig("pyplotexampleplot.png")

# Display the figure
plt.show()
```

```python
# Data for plotting
x = np.linspace(0.0, 1.0, 31)
f1 = x
f2 = x**2

# plotting
fig, ax = plt.subplots()
ax.plot(x, f1, color='r', marker='o', linestyle='-', label='f(x)=x')
ax.plot(x, f2, color='b', marker='x', linestyle='-.', label='f(x)=x^2')
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Our first plot!")

# draw the grid on the figure
ax.grid()

# create the legend
ax.legend()

# save the figure to a file you can then send to your friends
fig.savefig("ObjectOrientedexampleplot.png")

# display the figure
plt.show()
```

```python
N_points = 100000
n_bins = 20

# Generate a normal distribution centered at x=0
rng = np.random.default_rng()
x = rng.normal(0., 1., N_points)

# plt.subplot("number of rows", "number of columns", "subplot index")
plt.subplot(1,2,1)

# We can set the number of bins with the `bins` kwarg
plt.hist(x, bins=n_bins)
plt.ylabel("Count")
plt.xlabel("Data")

plt.subplot(1,2,2)
plt.hist(x, density=True, bins=n_bins)
plt.ylabel("Density")
plt.xlabel("Data")
```

```python
N_points = 100000
n_bins = 20

# Generate a normal distribution centered at x=0
rng = np.random.default_rng()
x = rng.normal(0., 1., N_points)

fig, axs = plt.subplots(1, 2, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[0].set_ylabel("Count")
axs[0].set_xlabel("Data")

axs[1].hist(x, density=True, bins=n_bins)
axs[1].set_ylabel("Density")
axs[1].set_xlabel("Data")
```
